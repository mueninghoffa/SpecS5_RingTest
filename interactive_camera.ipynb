```python
camera.color_mode = ueye.IS_CM_MONO12
camera.color_mode == ueye.IS_CM_MONO12
```

```python
camera.pixel_clock = min(camera.pixel_clock_possible_values)
camera.pixel_clock
```

```python
camera.exposure_time_ms = 1e3
camera.exposure_time_ms
```

```python
def take_img(exposure_time=None, subtract=True):
    if exposure_time is not None:
        camera.exposure_time_ms = exposure_time
        print(f"Exposure time set to {camera.exposure_time_ms}")

    _ = camera.expose()
    img = camera.read_numpy(subtract=subtract)
    fig = plt.figure(layout="constrained")
    plt.imshow(img)
    plt.colorbar()
    return img, fig
```

```python
img, fig = take_img(1e1, False)
```
